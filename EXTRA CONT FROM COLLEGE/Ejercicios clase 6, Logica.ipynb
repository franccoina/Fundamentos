##Ejemplo 4

Calcule el área de un triangulo y imprima su valor

Nota: el área de un triangulo es base*altura/2, tanto la base como la altura se le deben pedir al usuario, debe imprimir "El área del triangulo es : XXX"


Nota: si no se pide el valor numérico con la designación float(), este lo tomará como string (str)

```python
# Solicitar al usuario la longitud y el ancho de un triangulo

L = float(input("Favor ingresar la longitud de la base de su triangulo en cm"))
H = float(input("Favor ingresar la longitud de la altura de su triangulo en cm"))

# Calcular el área del triangulo

areaTriangulo = float(L*H/2)

# Imprimir el resultado

print("Para un triangulo con de base",L,"cm, y altura de",H,"cm, el area es igual a",areaTriangulo,"cm**2")
```

##Ejemplo 5

A partir de los numeros descritos en la variable (num1 y num2). Utiliza operadores relacionales para determinar si num1 es mayor que num2, si num1 es menor o igual que num2, y si num1 es igual a num2. Imprime los resultados booleanos de estas comparaciones (true or false).

```python
num1= 5
num2= 10

# Usar operadores relacionales

if num1>num2:
    print("es",True,"que",num1,"es mayor que",num2)
elif num1>num2:
    print("es",False,"que",num1,"es mayor que",num2)

if num1<num2:
    print("es",True,"que",num1,"es menor que",num2)
elif num1<num2:
    print("es",False,"que",num1,"es menor que",num2)


if num1==num2:
    print("es",True,"que",num1,"es igual que",num2)
elif num1==num2:
    print("es",False,"que",num1,"es igual que",num2)


# Imprimir los resultados
```

##Ejemplo 7

Utilice operadores relacionales en una expresión para determinar si x es menor que y y si x es igual a z. Imprime el resultado de la expresión en forma booleana (True or False).

```python
# Definición de variables
x = 5
y = 10
z = 5

# Usar operadores relacionales en una expresión

enunciado = x<y and x==z

# Imprimir el resultado de la expresión

print("El enunciado es",enunciado)
```

limite_inferior = 10

limite_superior = 20

```python
#Comprueba si la suma de dos números es mayor que el tercer número ingresado por el usuario.
num1 = float(input("Favor ingresar su primer numero"))
num2 = float(input("Favor ingresar su segundo numero"))
num3 = float(input("Favor ingresar su tercer numero"))

sumaNumeros = num1+num2

relacion = sumaNumeros>num3
relacion:True

print("Es",relacion,"que el resultado de la operacion",num1,"+",num2,", la cual es igual a",sumaNumeros,", es mayor que",num3)
```

```python
#Determina si un número ingresado por el usuario es positivo y par sin usar declaraciones condicionales.
numeroSolicitado = int(input("Favor ingresar su numero. Este debe ser Entero."))

numeroPar = (numeroSolicitado%2)==0
numeroPar:True

numeroPositivo = numeroSolicitado>0
numeroPositivo:True

respuesta = numeroPar and numeroPositivo

print("Es",respuesta,"que",numeroSolicitado,"es un numero Par y Positivo")
```

```python
#Verificar si un número está dentro de un rango específico
numeroSolicitado = float(input("Favor ingresar su numero"))

limite_inferior = 10
limite_superior = 20

rango = limite_inferior<numeroSolicitado<limite_superior

print("Es",rango,"que el",numeroSolicitado,"pertenece al rango entre",limite_inferior,"y",limite_superior)

```
